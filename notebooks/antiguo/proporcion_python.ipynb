# Análisis de Proporción de Género

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
file_path = '../data/raw/Paper1_WebData_Final.csv'
df = pd.read_csv(file_path)
print(df.head())
```

***ver cuantas mujeres y hombres hay***

```python
# Calcular la proporción de género
gender_counts = df['gender'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Proporción de Hombres y Mujeres (Edad 26-35)')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()
```

```python
# Calcular la frecuencia de cada categoría en 'NAICS_descr'
naics_counts = df['age_group'].value_counts()

# Crear el diagrama de torta
plt.figure(figsize=(10, 8))
plt.pie(
    naics_counts,
    labels=[f"{cat} ({pct:.1f}%)" for cat, pct in zip(naics_counts.index, naics_counts / naics_counts.sum() * 100)],
    autopct=None,
    startangle=140
)
plt.title('Distribución de age_group')
plt.axis('equal')
plt.show()
```

```python
categorias = ["L500k","L1k","L2k","L3k","L4k","L6k","L8k","R500k","R1k","R2k","R3k","R4k","R6k","R8k"]
# Lista de categorías que quieres graficar

# Calcular el promedio de cada categoría
promedios = df[categorias].mean()

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(promedios.index, promedios.values, color='skyblue')
plt.title('Promedio por categoría')
plt.ylabel('Valor promedio en Db')
plt.xlabel('Categoría')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
